    2. 为什么说Spring Boot 2.0易学难精？
        2.1 Spring Boot易学
            2.1.1 组件自动装配:规约大于配置，专注核心业务；
            2.1.2 外部化配置:一次构件、按需调配，到处运行（能通过外部配置来改变端口）；
            2.1.3 嵌入式容器:内置容器、无需部署、独立运行。
            2.1.4 Spring Boot Starter:简化依赖、按需装配、自我包含。
            2.1.5 Production-Ready: 一站式运维、生态无缝整合。
        2.2 Spring Boot难精
            2.2.1 组件自动装配:模式注解、@Enable模块、条件装配、加载机制；
            2.2.2 外部化配置:Environment抽象、生命周期、破坏性变更（Spring Boot1.0与2.0有些部分是不兼容的）；
            2.2.3 嵌入式容器:Servlet Web容器、Reactive Web容器；
            2.2.4 Spring Boot Starter:依赖管理、装配条件、装配顺序；
            2.2.5 Production-Ready:健康检查、数据指标、@Endpoint管控。
        2.3 Spring Boot与Java EE规范。
            2.3.1 Web:Servlet(JSR-315【3.0规范，包含了异步化】、JSR-340【4.0规范，包含了非阻塞】)
            2.3.2 SQL:JDBC(JSR-221[JDBC规范4.0])
            2.3.3 数据校验:Bean Validation(JSR 303【1.0版本】、JSR-349【1.2版本】)
            2.3.4 缓存:Java Caching API(JSR-107)
            2.3.5 WebSockets:Java API for WebSocket(JSR-356)
            2.3.6 Web Services:JAX-WS(JSR-224)
            2.3.7 Java管理:JMX（JSR3）
            2.3.8 消息:JMS(JSR-914)
